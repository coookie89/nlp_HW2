```python
import requests
import pandas as pd

from bs4 import BeautifulSoup
```

```python
MOVIE_ID=[]
MOVIE_TITLE=[]
CLASSIFICATION=[]
DESCRIPTION=[]
RELEASE_DATE=[]

def get_movie_info(movie_url): #爬電影資訊
    response = requests.get(movie_url) 
    response.encoding = 'UTF-8'
    soup = BeautifulSoup(response.text, "lxml")
    
    informations=soup.find("div", id="content_l")
    
    #print("電影名稱: ")
    movie_title=informations.find("h1").getText()
    MOVIE_TITLE.append(movie_title) #把電影名稱寫入list
    print(movie_title)
    #print()
    
    classifications=informations.find("div", class_="level_name_box")
    classifications=classifications.find_all("div", class_="level_name")
    #print(classifications)
    #print("類別: ")
    classification_list=[]
    for classification in classifications:
        classification=classification.find("a").getText()
        
        classification=classification.replace("\n","")
        classification=classification.replace(" ","")
        classification_list.append(classification)
        
        #print(classification)
        
    if classification_list is not None:
        CLASSIFICATION.append(classification_list) #把電影類別寫入list
    else:
        CLASSIFICATION.append("NaN")
    
    #print("劇情介紹: ")
    description=informations.find(id="story").getText()
    description=description.replace("\n            ","")
    description=description.replace("\r","")
    DESCRIPTION.append(description) #把電影介紹寫入list
    #print(description.replace(" ",""))
    #print()
    
    #print("上映日期: ")
    release_date=informations.find_all("span")[0].getText()
    RELEASE_DATE.append(release_date[5:]) #把上映日期寫入list
    #print(release_date[5:])
    #print()


def csv_output(batch, movie_id, movie_name, classification, description, release_date):
    data = {
        '電影id': movie_id,
        '名稱': movie_name,
        '分類': classification,
        '劇情介紹': description,
        '上映日期' : release_date
    }

    movie_info = pd.DataFrame(data) #建立DataFrame物件
    if batch==0: #如果為第一次寫入
        movie_info.to_csv('movie_info.csv', encoding="utf-8", index=0) #寫入csv檔
    else:
        movie_info.to_csv('movie_info.csv', encoding="utf-8", mode='a', header=False, index=0) #寫入csv檔

    #print(movie_info)
    
def txt_output(element, file_name):
    file = open(file_name, 'a')
    file.write(element+"\n")
    file.close()
    
def txt_to_list(file_path):
    lines = []
    with open(file_path, 'r', encoding="UTF8") as file:
        while True:
            line = file.readline()
            if not line:
                break
            line = line.strip('\n')
            lines.append(line)
    return lines
    

if __name__ == "__main__":
    id_start=1 #從第幾個id開始抓
    id_finish=11466 #要抓取的id範圍
    
    data_num_count=0 #計算抓取的資料數目
    batch_num=100 #幾筆寫入csv一次
    status=0 #紀錄是不是第一次寫入
    
    movie_already_crawl=txt_to_list("movie_id_already_crawl.txt") #已經爬過的movie_id
    
    for movie_id in range(id_start,id_finish+1):
        movie_url="https://movies.yahoo.com.tw/movieinfo_main.html/id="+str(movie_id)
        
        try:
            if movie_id not in movie_already_crawl:
                get_movie_info(movie_url)
                
                data_num_count=data_num_count+1
                print(data_num_count," id-",movie_id)

                MOVIE_ID.append(movie_id)
                movie_already_crawl.append(movie_id)
                txt_output(str(movie_id), "movie_id_already_crawl.txt")

                if data_num_count%batch_num==0:
                    if status==0: #如果是第一次寫入
                        csv_output(0, MOVIE_ID, MOVIE_TITLE, CLASSIFICATION, DESCRIPTION, RELEASE_DATE)
                        status=1
                    else:
                        csv_output(1, MOVIE_ID, MOVIE_TITLE, CLASSIFICATION, DESCRIPTION, RELEASE_DATE)

                    MOVIE_ID.clear()
                    MOVIE_TITLE.clear()
                    CLASSIFICATION.clear()
                    DESCRIPTION.clear()
                    RELEASE_DATE.clear()
            
        except:
            continue
            
    
    if MOVIE_TITLE is not None:
        csv_output(1, MOVIE_ID, MOVIE_TITLE, CLASSIFICATION, DESCRIPTION, RELEASE_DATE)
        
    print ("這回合結束ㄌ")
```
